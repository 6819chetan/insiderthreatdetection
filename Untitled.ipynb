```python
import pandas as pd
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('merged_all_normalized.csv')
```

```python
df
```

```python
X = df.drop(['user', 'date'], axis=1)
```

```python
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.005)  
anomalies = lof.fit_predict(X)
```

```python
anomalies = np.where(anomalies == -1, -1, 1)
```

```python
df['anomaly'] = anomalies
```

```python
cluster_counts = df['anomaly'].value_counts()
print("Overall cluster counts:")
print(cluster_counts)
```

```python
unique_users_per_cluster = df.groupby('anomaly')['user'].nunique()
print("Unique user counts per cluster:")
print(unique_users_per_cluster)
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=anomalies, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Local Outlier Factor Anomaly Detection')
plt.show()
```
